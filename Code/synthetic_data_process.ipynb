# Synthetic Data Processing

```python
##import necessary variables
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
## read synthetic data using pandas
data = pd.read_csv('generated_samples.csv')
data.info()
```

```python
# print first 5 rows
data.head()
```

```python
##see the columns
data.columns
```

```python
##only select columns that has numeric data type
columns = ['TemP (K)', 'Time (min)', 'PS (mm)', 'BET (m2/g)', 'PV (cm3)',
       'C (wt.%)', 'H  (wt.%)', 'N  (wt.%)', 'O  (wt.%)', 'Qm (mg/g)']
```

```python
## Let's only keep positive values since these values are supposed to be positive
data_clean = data[data['TemP (K)']> 0]
data_clean = data[data['Time (min)']>0]
data_clean = data[data['PS (mm)']>0]
data_clean = data[data['BET (m2/g)']>0]
data_clean = data[data['PV (cm3)']>0]
data_clean = data[data['C (wt.%)']>0]
data_clean = data[data['H  (wt.%)']>0]
data_clean = data[data['N  (wt.%)']>0]
data_clean = data[data['O  (wt.%)']>0]
data_clean = data[data['Qm (mg/g)']>0]
```

```python
data_clean.head()
```

```python
## another dataframe just for numerical columns
first_df = data_clean[columns]
```

```python
##Let's select columns that have non-numeric values and supposed to be binary
tp_col = ['TP_Benzocaine', 'TP_Ciprofloxacin', 'TP_Citalopram', 'TP_Diclofenac ',
       'TP_Dimetridazole', 'TP_Floxentine', 'TP_Metronidazole',
       'TP_Nitroimidazole', 'TP_Norfloxacin', 'TP_Oxytetracycline',
       'TP_Oxytetracycline ', 'TP_Salicylic acid', 'TP_Sulfadiazine',
       'TP_Sulfamethoxazole', 'TP_Tetracycline', 'TP_Triclosan',
       'TP_ibuprofen', 'TP_sulfamethazine ']
```

```python
# one dataframe for one hot encoded variables
tp_data = data_clean[tp_col]
tp_data.info()
```

```python
# let's create a dataframe sime size as our synthetic data with only with 0's
df_binary = pd.DataFrame(0, index=tp_data.index, columns=tp_data.columns)

# Find the column with the maximum value for each row
max_cols = tp_data.idxmax(axis=1)

# Set the maximum value to 1 for each row in the corresponding column
for i, col in enumerate(max_cols):
    df_binary.at[i, col] = 1
```

```python
##concat this new dataframe with with numeric dataframe
processed_data = pd.concat([first_df, df_binary], join = 'inner', axis=1)
```

```python
processed_data.info()
```

```python
##not necessary but we did sort dataframe based on target variable
df_sorted =processed_data.sort_values(by = 'Qm (mg/g)')
```

```python
##Box plot for the entire dataframe
df_sorted[columns].plot(kind='box',figsize=(12,6),subplots=True, layout = (2,5))
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
plt.suptitle('Boxplot for each features for synthetic data', fontsize=16)
plt.savefig('boxplot_synthetic_data.png', dpi = 300)
plt.show()
```

```python
##Since we have generated much more data than our requirements and it has computation cost we decided to
# take 5 percent of the dataset by taking random sample.
sampled_df = df_sorted.sample(frac=0.05)
sampled_df.info()
```

```python
#write the dagta as csv file
sampled_df.to_csv('synthetic_data.csv', index = False)
```

```python
n_columns = 10

# Read the Actual data
Actual_data = pd.read_excel('updated_DATASET.xlsx', engine="openpyxl", usecols=range(n_columns))

# Read the synthetic data
synthetic_data = pd.read_csv('synthetic_data.csv', usecols=range(n_columns))
```

```python
# Create a column to indicate the type of data
Actual_data['Type'] = 'Actual'
synthetic_data['Type'] = 'Synthetic'

# Concatenate the two datasets
combined_data = pd.concat([Actual_data, synthetic_data])

# Melt the data for easier plotting
melted_data = pd.melt(combined_data, id_vars='Type', var_name='Column', value_name='Value')
```

```python
import seaborn as sns
# Set the aesthetic style of the plots
sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})

# Create a grid of subplots
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(20, 10))  # Adjust the layout according to the number of columns
axes = axes.flatten()  # Flatten the array of axes to easily index them

# Plot each column in a separate subplot
for i, column in enumerate(melted_data['Column'].unique()):
    sns.boxplot(
        x='Type',
        y='Value',
        data=melted_data[melted_data['Column'] == column],
        ax=axes[i],
        notch=True
    )
    axes[i].set_title(column, fontsize=12)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Value' if i % 5 == 0 else '')  # Display y-axis label only on the leftmost plots for clarity
    axes[i].tick_params(labelsize=10)

# Adjust the layout
plt.savefig('Boxplot_comparison_org_syn.png', dpi = 300)
plt.tight_layout(pad=2.0)

# Show the plot
```
